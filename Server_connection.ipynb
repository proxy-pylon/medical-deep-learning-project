```python
# example usage
import paramiko, getpass
from scp import SCPClient

host = "89.169.111.191"
user = "rakhim"
passwd = getpass.getpass("Server password: ")  # avoids echoing password

ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
ssh.connect(hostname=host, username=user, password=passwd)
print("✅ Connected to server")
scp = SCPClient(ssh.get_transport())
scp.put("Updated_Model.ipynb", "~/csci494-final-project/") #need to upload the file into the colab
scp.close()
print("✅ Notebook uploaded to server")

stdin, stdout, stderr = ssh.exec_command('ls -la ~/csci494-final-project')
print(stdout.read().decode())
ssh.close()
```

```python
import paramiko

ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
ssh.connect(hostname=host, username=user, password=passwd)
print("✅ Connected to server")

# Create and activate venv + install dependencies
commands = [
    "python3 -m venv ~/papermill_venv",
    "source ~/papermill_venv/bin/activate && pip install --quiet --upgrade pip && pip install --quiet papermill jupyter ipykernel kagglehub",
    "source ~/papermill_venv/bin/activate && python -m ipykernel install --user --name=papermill_venv",
    # Install project requirements if available
    "if [ -f ~/csci494-final-project/requirements.txt ]; then source ~/papermill_venv/bin/activate && pip install --quiet -r ~/csci494-final-project/requirements.txt; fi",
]

for cmd in commands:
    print(f"\n→ {cmd}")
    stdin, stdout, stderr = ssh.exec_command(cmd)
    for line in iter(stdout.readline, ""):
        print(line, end="")
    err = stderr.read().decode()
    if err:
        print(err)

# Execute notebook with live streaming (real-time output)
print("\n→ Executing notebook with live output...\n")

command = (
    "source ~/papermill_venv/bin/activate && "
    "papermill --progress-bar --log-output "
    "~/csci494-final-project/Updated_Model.ipynb "
    "~/csci494-final-project/Updated_Model_executed.ipynb"
)
stdin, stdout, stderr = ssh.exec_command(command)

# Print output in real time
for line in iter(stdout.readline, ""):
    print(line, end="")

# Capture any final errors
errors = stderr.read().decode()
if errors.strip():
    print("\n❌ Errors:\n", errors)
else:
    print("\n✅ Notebook executed successfully")

ssh.close()
```
